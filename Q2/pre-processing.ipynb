## Enrollment-Grade Roster

```python
import pandas as pd
import re

# Example: your DataFrame
# df = pd.DataFrame({'Subject Code / Name': [...]})

def extract_subject_code(row):
    text = str(row).strip()

    # Case 1: Looks like "IT 989/20 / Thesis / Research hours"
    match = re.match(r'^([A-Z]{2,}\s?\d+/\d+)', text)
    if match:
        return match.group(1)

    # Case 2: Looks like "GNL 102 / Physical Education 1"
    # Only remove space if / has space on both sides
    return re.sub(r'\s+/\s+', '/', text)

# Create new column
grade['Subject Code / Name (Cleaned)'] = grade['Subject Code / Name'].apply(extract_subject_code)
grade=grade.drop('Subject Code / Name',axis=1)
grade = grade.rename(columns={'Subject Code / Name (Cleaned)': 'Subject Code / Name'})
```

```python
import pandas as pd

def clean_course(course):
    # Remove bracketed part if it contains any digit
    if '(' in course and ')' in course:
        start = course.rindex('(')
        end = course.rindex(')')
        if start < end:  # ensure it's a valid pair
            inside = course[start+1:end]
            if any(c.isdigit() for c in inside):
                course = course[:start] + course[end+1:]

    # Remove specific prefixes
    if course.startswith("T1-24-25-"):
        course = course[len("T1-24-25-"):]
    elif course.startswith("T2-24-25-"):
        course = course[len("T2-24-25-"):]

    return course.strip()


attendance["Course"] = attendance["Course"].apply(clean_course)
```

```python
def clean_slash_spacing(course):
    if '/' in course:
        # Check for space around slash and remove it
        parts = course.split('/')
        if len(parts) == 2:
            left = parts[0].rstrip()
            right = parts[1].lstrip()
            return left + '/' + right
    return course

attendance["Course"] = attendance["Course"].apply(clean_slash_spacing)
```

```python
def remove_by_and_after(course):
    if " by " in course:
        course = course.split(" by ")[0]
    return course.strip()

# Apply to DataFrame
attendance["Course"] = attendance["Course"].apply(remove_by_and_after)
```

```python
def replace_hyphen_with_slash(course):
    if '-' in course:
        hyphen_index = course.find('-')
        if hyphen_index > 0 and course[hyphen_index - 1].isdigit() or course[hyphen_index - 1] == ' ':
            course = course[:hyphen_index] + '/' + course[hyphen_index + 1:]
    return course

# Apply to DataFrame
attendance["Course"] = attendance["Course"].apply(replace_hyphen_with_slash)
```

## Joining with the merged_df to create the final fact table

```python
mapping = {
    "AIM 511 Machine Learning": "AIM 511/Machine Learning",
    "DHS 304 User Research": "DHS 304/User Research",
    "DHS 301 Data Analysis and Visualization": "DHS 301/Data Analysis and Visualization",
    "GNL 801 Optimization": "GNL 801/Optimization",
    "NWC 882/Special Topics / Network-Based Computing for HPC": "NWC 882/Special Topics - Network-Based Computing for HPC",
    "EGC 103/Introdution to Computer Architecture and Operating Systems": "EGC 103/Introduction to Computer Architecture and Operating Systems",
    "AI 835 Self-Supervised Learning":"AI 835/Self-Supervised Learning",
    "DT 311 Software Product Management":"DT 311/Software Product Management",
    "EGC 223 /Computer Architecture - Memory":"EGC 223/Computer Architecture - Memory",
    "AI 835 Self-Supervised Learning":"AI 835/Self-Supervised Learning"
}

attendance['Course'] = attendance['Course'].apply(lambda x: mapping.get(x, x))
grade['Subject Code / Name'] = grade['Subject Code / Name'].apply(lambda x: mapping.get(x, x))
enrollment['Subject Code / Name'] = enrollment['Subject Code / Name'].apply(lambda x: mapping.get(x, x))
```

```python
import pandas as pd

# Inner join on 'Member Id' and 'Course' from attendance with 
# 'Student ID' and 'Subject Code / Name' from grade
merged = pd.merge(attendance, grade, 
                  left_on=["Member Id", "Course"], 
                  right_on=["Student ID", "Subject Code / Name"],
                  how="inner")

# Group by the join keys and aggregate the desired columns
fact_table = merged.groupby(["Member Id", "Course"], as_index=False).agg({
    "Number of classes attended": "sum",
    "Number of classes absent": "sum",
    "Course Credit": "max",
    "Average Attendance %":"max"
})

# Output the fact table to CSV in Kaggle working directory
fact_table.to_csv("/kaggle/working/fact_table_final1.csv", index=False)
grade.to_csv("/kaggle/working/grade.csv", index=False)
attendance.to_csv("/kaggle/working/attendance.csv", index=False)
enrollment.to_csv("/kaggle/working/enrollment.csv", index=False)
```

```python
# --- Step 4: Merge the fact table with the reduced enrollment data ---
# This merge will filter fact_table, ensuring only rows that have matching enrollment data are present.
fact_table_final = pd.merge(
    fact_table, enrollment,
    left_on=["Member Id", "Course"],
    right_on=["Student ID", "Subject Code / Name"],
    how="inner"
)

# If you want to drop the extra enrollment join columns from the final DataFrame:
fact_table_final.drop(columns=["Student ID", "Subject Code / Name"], inplace=True)

# --- Optionally, output to CSV ---
fact_table_final.to_csv("/kaggle/working/fact_table_final2.csv", index=False)
```
